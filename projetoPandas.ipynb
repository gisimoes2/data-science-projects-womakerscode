```python
horas_sono_eng_media = df.groupby('Profissão')['Duração do sono'].mean()
# horas_sono_eng_moda = df.groupby('Profissão')['Duração do sono'].mode()
horas_sono_eng_mediana = df.sort_values(['Profissão','Duração do sono'])
horas_sono_eng_mediana = horas_sono_eng_mediana.groupby('Profissão')['Duração do sono'].median()
print(horas_sono_eng_media)
print(horas_sono_eng_mediana)
# print(horas_sono_eng_moda)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from mpl_toolkits.mplot3d import Axes3D
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
saude = pd.read_csv('saude (1).csv')
saude
```

```python
saude.head()
```

```python
saude.columns
```

```python
idade = saude['age']
id = saude['id']
plt.bar(id,idade)
plt.xlabel('Id')
plt.ylabel('Idade')
plt.title('Gráfico de idades ')
plt.show()
```

```python
sns.scatterplot(data= saude , x="height",y="weight") #hue é a legenda
plt.title('Gráfico da relação entre peso e altura')
plt.show
#sns.get_dataset_names()
```

```python

fig = px.bar(saude, x='id', y='gender', color='sex', barmode='group', title= 'Pessoas e genero')
fig.show()
```

```python
vendas =  pd.read_csv('vendas (1).csv')
vendas
```

```python
vendas.head()
```

```python
#mudando o conteudo da coluna date para os dias do mes de janeiro
vendas['date'] = pd.to_datetime(vendas['date'])

# Alterar a coluna 'date' para exibir apenas os dias
vendas['date'] = vendas['date'].dt.day
```

```python
vendas.columns
```

```python
vendas.head()
```

```python
preco = vendas['sales']
tempo = vendas['date']
plt.plot(tempo,preco)
plt.xlabel('janeiro de 2023')
plt.ylabel('vendas')
plt.title('Vendas ao longo do tempo')
plt.show()
```

2. Qual é a média e a mediana para cada uma das disciplinas? (Lembre-se
de remover todos os valores nulos quando considerar a mediana)

```python
#Ordendando os valores para calcular a mediana
df.sort_values('Linguagens')
df.sort_values('Ci\u00eancias humanas')
df.sort_values('Ci\u00eancias da natureza')
df.sort_values('Matem\u00e1tica')
df.sort_values('Reda\u00e7\u00e3o')

#Calcular a mediana
df['Linguagens'].median()
df['Ci\u00eancias humanas'].median()
df['Ci\u00eancias da natureza'].median()
df['Matem\u00e1tica'].median()
df['Reda\u00e7\u00e3o'].median()

#obtendo os valores para calcular a média
soma_ling = df['Linguagens'].sum()
soma_ch =  df['Ci\u00eancias humanas'].sum()
soma_cn = df['Ci\u00eancias da natureza'].sum()
soma_mat = df['Matem\u00e1tica'].sum()
soma_red = df['Reda\u00e7\u00e3o'].sum()

quant_ling = df['Linguagens'].count()
quant_ch = df['Ci\u00eancias humanas'].count()
quant_cn = df['Ci\u00eancias da natureza'].count()
quant_M = df['Matem\u00e1tica'].count()
quant_R = df['Reda\u00e7\u00e3o'].count()

#calcular a média
print(soma_ling/quant_ling)
print(soma_ch/quant_ch)
print(soma_cn/quant_cn)
print(soma_mat/quant_M)
print(soma_red/quant_R)

#ou
media_ling = df['Linguagens'].mean()
df['Ci\u00eancias humanas'].mean()
df['Ci\u00eancias da natureza'].mean()
media_mat = df['Matem\u00e1tica'].mean()
df['Reda\u00e7\u00e3o'].mean()
```

3. Considerando o curso de Ciência da Computação da UFPE, onde o peso
cada uma das disciplinas ponderado:
a. Redação - 2
b. Matemática e suas Tecnologias - 4
c. Linguagens, Códigos e suas Tecnologias - 2
d. Ciências Humanas e suas Tecnologias - 1
e. Ciências da Natureza e suas Tecnologias - 1
Qual o desvio padrão e média das notas dos 500 estudantes mais bem
colocados considerando esses pesos?


```python
primeiras_500_linhas = df.head(500)
pesos = {
    'Redação': 2,
    'Matemática': 4,
    'Linguagens': 2,
    'Ciências Humanas': 1,
    'Ciências da Natureza': 1
}

nota_ponderada = (df(['Reda\u00e7\u00e3o'] * pesos['Redação']) + df(['Matem\u00e1tica'] * pesos['Matemática']) +
df(['Linguagens'] * pesos['Linguagens']) + df(['Ci\u00eancias humanas'] * pesos['Ciências Humanas'])
+ df(['Ci\u00eancias da natureza'] * pesos['Ciências da Natureza'])) /sum(pesos.values())

media_ponderada = nota_ponderada.mean()
print(media_ponderada)
desvio_padrao = nota_ponderada.std()
print(desvio_padrao)
```
